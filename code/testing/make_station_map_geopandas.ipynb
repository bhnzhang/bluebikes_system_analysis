```python
# get dataframe containing one day of logged data

# read from json files
# log_220715 = load_jsons_savetocsv( 'C:\\Users\\beezy\\git\\bluebikes_system_analysis\\data\\station data 220715 jsons', 
#                                     'log_220715.csv' )
# log_220715 = load_jsons_toDF( 'C:\\Users\\beezy\\git\\bluebikes_system_analysis\\data\\station data 220715 jsons' )

# read from csv
filetoload = 'C:\\Users\\beezy\\git\\bluebikes_system_analysis\\data\\station data 220715 jsons\\log_220715.csv' 
log_220715 = pd.read_csv(filetoload, index_col=0) # read csv into df

# drop any duplicates
log_220715 = log_220715.drop_duplicates()
```

```python
log_220715
```

```python
# extract hour and minute from timestamps

# convert time to datetime
log_220715['last_reported'] = pd.to_datetime( log_220715['last_reported'], unit='s' )

# add hour, minute column
log_220715['hour'] = log_220715['last_reported'].dt.hour
log_220715['minute'] = log_220715['last_reported'].dt.minute
```

```python
log_220715
```

```python
# only keep entries at a selected hour and minutes
selected_hour     = 8
selected_min_low  = 0
selected_min_high = 59
log_220715_crop = log_220715.loc[ log_220715['hour'] == selected_hour ]
log_220715_crop = log_220715_crop.loc[ ( log_220715_crop['minute'] >= selected_min_low ) & 
                             ( log_220715_crop['minute'] <= selected_min_high ) ]
log_220715_crop
```

```python
# convert to a geodataframe and plot on explorer map
station_gdf = geopandas.GeoDataFrame(
    log_220715_crop, geometry=geopandas.points_from_xy(log_220715_crop['lon'], log_220715_crop['lat']), crs='EPSG:4326')

# i think i have to remove timestamp column before plotting?
station_gdf = station_gdf.drop('last_reported', axis=1)
station_gdf
```

```python
# i want to find entries with unreal values
import numpy as np
station_gdf = station_gdf.replace('', np.nan) # this replaces empty strings with NaN quickly
[rows2drop, cols2drop] = np.where(pd.isnull(station_gdf)) # this uses numpy to find NaN values
indices2drop = station_gdf.index[rows2drop]
indices2drop
```

```python
# convert time to datetime
log_220715['last_reported'] = pd.to_datetime( log_220715['last_reported'], unit='s' )
```

```python
# calculate current capacity and add that as a column
log_220715['current_capacity'] = log_220715['num_bikes_available']/log_220715['capacity']
```

```python
# add hour column
log_220715['hour'] = log_220715['last_reported'].dt.hour
log_220715
```

```python
# pick an hour to select

# first group on station id and hour
log_220715_groupedby_id_hour = log_220715.groupby(['station_id','hour'])
log_220715_groupedby_id_hour
```

```python
# get mean and std
# mean_std_groupedby_id_hour = log_220715_groupedby_id_hour.agg(['mean','std']) # for all columns
mean_std_groupedby_id_hour = log_220715_groupedby_id_hour.agg(['mean']) # for all columns
mean_std_groupedby_id_hour

# this might be necessary for some reason
mean_std_groupedby_id_hour = mean_std_groupedby_id_hour.reset_index()
mean_std_groupedby_id_hour
```

```python
# cut down the table to just the hour of interest
hour_to_plot = 8
indx_wantedhour = mean_std_groupedby_id_hour['hour'] == hour_to_plot

data_hourtoplot = mean_std_groupedby_id_hour.loc[indx_wantedhour]
data_hourtoplot
```

```python
# convert panda df to geodf

# add lat and lon as columns
data_hourtoplot['lonflat'] = data_hourtoplot['lon']['mean']
data_hourtoplot['latflat'] = data_hourtoplot['lat']['mean']

station_gdf = geopandas.GeoDataFrame(
    data_hourtoplot, geometry=geopandas.points_from_xy(data_hourtoplot['lonflat'], data_hourtoplot['latflat']), crs='EPSG:4326')
station_gdf
```

```python
# add mean capacity as a column
station_gdf['mean_capacity'] = station_gdf['current_capacity']['mean']
station_gdf.columns
# gdf = station_gdf.copy()
```
